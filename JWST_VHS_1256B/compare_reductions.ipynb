```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.io import fits
```

```python
fig, ax = plt.subplots(figsize=(12,4))
for filename in ['nirspec_140h_1.dat', 'nirspec_140h_2.dat', 'nirspec_235h_1.dat']:

    wave, flux, err = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/reduction_31_07_2025/{filename}', delimiter=',').T
    wave_M, flux_M, err_M = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/ERS_reduction/{filename}', delimiter=',').T
    # wave_M, flux_M, err_M = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/reduction_05_03_2025/{filename}', delimiter=',').T

    ax.plot(wave_M, flux_M, c='indianred', lw=0.5, label='ERS')
    ax.plot(wave, flux, c='k', lw=0.8, label='New reduction')

    ax.plot(wave, flux-np.interp(wave, wave_M, flux_M), c='k', lw=0.8, alpha=0.5, label='Diff')

ax.legend(labelcolor='linecolor')
ax.set(xlabel='Wavelength [um]', ylabel='Flux [W/m^2/um]')#, xlim=(1.7,1.8))
ax.axhline(0, c='k', lw=0.8)
plt.tight_layout()
plt.savefig('./g140h.pdf', dpi=300)
plt.show()

fig, ax = plt.subplots(figsize=(12,4))
for filename in ['nirspec_235h_1.dat', 'nirspec_235h_2.dat']:

    wave, flux, err = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/reduction_31_07_2025/{filename}', delimiter=',').T
    wave_M, flux_M, err_M = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/ERS_reduction/{filename}', delimiter=',').T
    # wave_M, flux_M, err_M = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/reduction_05_03_2025/{filename}', delimiter=',').T

    ax.plot(wave_M, flux_M, c='indianred', lw=0.5, label='ERS')
    ax.plot(wave, flux, c='k', lw=0.8, label='New reduction')

    ax.plot(wave, flux-np.interp(wave, wave_M, flux_M), c='k', lw=0.8, alpha=0.5, label='Diff')

ax.legend(labelcolor='linecolor')
ax.set(xlabel='Wavelength [um]', ylabel='Flux [W/m^2/um]')
ax.axhline(0, c='k', lw=0.8)
plt.tight_layout()
plt.savefig('./g235h.pdf', dpi=300)
plt.show()

fig, ax = plt.subplots(figsize=(12,4))
for filename in ['nirspec_395h_1.dat', 'nirspec_395h_2.dat']:

    wave, flux, err = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/reduction_31_07_2025/{filename}', delimiter=',').T
    wave_M, flux_M, err_M = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/ERS_reduction/{filename}', delimiter=',').T
    # wave_M, flux_M, err_M = np.loadtxt(f'/net/schenk/data2/regt/JWST_reductions/Miles_ea_2023/reduction_05_03_2025/{filename}', delimiter=',').T

    ax.plot(wave_M, flux_M, c='indianred', lw=0.5, label='ERS')
    ax.plot(wave, flux, c='k', lw=0.8, label='New reduction')

    ax.plot(wave, flux-np.interp(wave, wave_M, flux_M), c='k', lw=0.8, alpha=0.5, label='Diff')

ax.legend(labelcolor='linecolor')
ax.set(xlabel='Wavelength [um]', ylabel='Flux [W/m^2/um]')
ax.axhline(0, c='k', lw=0.8)
plt.tight_layout()
plt.savefig('./g395h.pdf', dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,10), nrows=3)
for i, filename in enumerate(['miri_1A.dat', 'miri_1B.dat', 'miri_1C.dat', 'miri_2A.dat', 'miri_2B.dat', 'miri_2C.dat', 'miri_3A.dat', 'miri_3B.dat', 'miri_3C.dat', 'miri_4A.dat']):

    wave, flux, err = np.loadtxt(f'./Miles_ea_2023/reduction_05_03_2025/{filename}', delimiter=',').T
    wave_M, flux_M, err_M = np.loadtxt(f'./Miles_ea_2023/ERS_reduction/{filename}', delimiter=',').T

    ax[0].plot(wave_M, flux_M, c='indianred', lw=0.5, label={'0':'ERS'}.get(str(i)))
    ax[0].plot(wave, flux, c='k', lw=0.8, label={'0':'New reduction'}.get(str(i)), alpha=0.5)

    ax[1].plot(wave_M, flux_M, c='indianred', lw=0.8, label={'0':'ERS'}.get(str(i)))
    ax[2].plot(wave, flux, c='k', lw=0.8, label={'0':'New reduction'}.get(str(i)))

ax[1].legend(); ax[2].legend()
ax[1].set(xlim=(7,12), ylim=(0,1e-16))
ax[2].set(xlabel='Wavelength [um]', xlim=(7,12), ylim=(0,1e-16))

plt.show()
```
